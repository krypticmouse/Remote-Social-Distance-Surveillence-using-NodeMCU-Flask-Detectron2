```
from flask import Flask, Response, render_template
from get_frames_yolo import gen_frames
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)

@app.route('/')
def main():
    return render_template('index.html')

@app.route('/video1_feed')
def video1_feed():
    return Response(gen_frames('people.mp4',1), mimetype='multipart/x-mixed-replace; boundary=frame')

@app.route('/video2_feed')
def video2_feed():
    return Response(gen_frames('people.mp4',2), mimetype='multipart/x-mixed-replace; boundary=frame')

@app.route('/video3_feed')
def video3_feed():
    return Response(gen_frames('people.mp4',3), mimetype='multipart/x-mixed-replace; boundary=frame')

@app.route('/video4_feed')
def video4_feed():
    return Response(gen_frames('people 3.mp4',4), mimetype='multipart/x-mixed-replace; boundary=frame')

if __name__ == '__main__':
    app.run()
```
